```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('water_potability.csv')
```

```python
data.head()
```

```python
data['Turbidity'].isna().sum()
```

```python
data.describe()
```

```python
sns.distplot(data['Turbidity'],bins=25)
```

```python
std = data['Turbidity'].std()
mean = data['Turbidity'].mean()
q1 = data['Turbidity'].quantile(0.25)
q2 = data['Turbidity'].median()
q3 = data['Turbidity'].quantile(0.75)
```

### Detecting Outliers using Inter quartile range

```python
sns.boxplot(y='Turbidity',data=data)
```

```python
IQR = q3-q1
lower_limit = q1 - 1.5*IQR
upper_limit = q3 + 1.5*IQR
```

```python
data[(data['Turbidity'] > upper_limit) | (data['Turbidity'] < lower_limit)][['Turbidity']]
```

### Detecting Outliers using 3 Std Dev

#### As we can see in the above image 
- 68.26 % data lies within the first standard deviation
- 95.44 % data lies within the second standard deviation
- 99.62 % data lies within the third standard deviation

#### The rest of the data are considered as outliers i.e. in most of the cases 3 standard dev is taken to detect the outliers

```python
upper_limit = mean + 3*std
lower_limit = mean - 3*std
data[(data['Turbidity'] > upper_limit) | (data['Turbidity'] < lower_limit)][['Turbidity']]
```

### Q1 and Q3 in terms of how many standard deviations away they are from the mean (z-score)

#### So we basically need to find out the values of Q1 and Q3 in terms of standard deviation i.e. how many standard deviations away they are from the mean. We do this using the Z-score

```python
z1 = (q1 - mean)/std
z2 = (q3 - mean)/std
```

#### For value 1 ( The normal data points will be considered as outliers )

```python
print((z1 - 1 * (z2-z1)*std)/std,'Std')
```

```python
print((z2 + 1 * (z2-z1)*std)/std,'Std')
```

####  For value 1.5 (Just the right threshold, close to 3 std deviation)

```python
print((z1 - 1.5 * (z2-z1)*std)/std,'Std')
```

```python
print((z2 + 1.5 * (z2-z1)*std)/std,'Std')
```

#### For value 2 ( The outliers will be considered as normal data points)

```python
print((z1 - 2 * (z2-z1)*std)/std,'Std')
```

```python
print((z2 + 2 * (z2-z1)*std)/std,'Std')
```

**Conclusion:** We can see that the value 1.5 gives us the 3 standard deviations away equivalent unlike the values 1 or 2 which will either consider the outliers as normal data points or even consider the normal data points as outliers.
